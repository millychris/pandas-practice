```python
import pandas as pd
import sqlite3
```

```python
# 建立連線
con = sqlite3.connect('../data/test_db.sqlite')
```

```python
# 把 311 資料寫進資料庫
requests = pd.read_csv('../data/311-service-requests.csv', dtype='unicode')
requests.to_sql('requests', con, if_exists='replace', index=False)
```

```python
pd.read_sql('SELECT * FROM requests LIMIT 5', con)
```

```python
pd.read_sql('SELECT "Complaint Type", "Borough" FROM requests WHERE "Borough" = \'BROOKLYN\' LIMIT 10', con)
```

```python
pd.read_sql('''
    SELECT "Borough", COUNT(*) as complaint_count 
    FROM requests 
    GROUP BY "Borough" 
    ORDER BY complaint_count DESC
''', con)
```
