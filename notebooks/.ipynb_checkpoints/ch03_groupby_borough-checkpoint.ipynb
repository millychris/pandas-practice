```python
import pandas as pd
```

```python
complaints = pd.read_csv('../data/311-service-requests.csv', dtype='unicode')
```

```python
noise_complaints = complaints[complaints['Complaint Type'].str.contains('Noise')]
noise_complaints['Borough'].value_counts()
```

```python
noise_complaints.groupby('Borough').size()
```

```python
complaints.groupby(['Borough', 'Complaint Type']).size().sort_values(ascending=False)[:20]
```

```python
noise_complaints.groupby(['Borough', 'Complaint Type']).size().unstack().fillna(0)
```
