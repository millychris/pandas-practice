```python
import pandas as pd
```

```python
weather_2012 = pd.read_csv('../data/weather_2012.csv', parse_dates=True, index_col='Date/Time',dtype='unicode')
weather_2012[:5]
```

```python
weather_2012.shape
```

```python
weather_2012['Weather'].str.len()[:5]
```

```python
weather_description = weather_2012['Weather']
weather_description[:10]
```

```python
is_snowing = weather_description.str.contains('Snow', regex=False, na=False)
is_snowing[:10]
```

```python
is_snowing = is_snowing.astype(float)
is_snowing.plot()
```

```python
is_snowing.resample('ME').sum().plot(kind='bar')
```
