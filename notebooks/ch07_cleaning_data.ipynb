```python
import pandas as pd
```

```python
requests = pd.read_csv('../data/311-service-requests.csv', dtype='unicode')
requests[:5]
```

```python
# 每欄有幾個缺失值
requests.isnull().sum()
```

```python
# 看看郵遞區號有哪些奇怪的值
requests['Incident Zip'].unique()
```

```python
# 把明顯錯誤的值換成 NaN
import numpy as np
```

```python
na_values = ['NO CLUE', 'N/A', '0']
requests['Incident Zip'] = requests['Incident Zip'].replace(na_values, np.nan)
#  np.nan 是浮點數，可以直接參與數學運算而不會報錯——比如 np.nan + 1 結果還是 np.nan，不會崩潰。None 是 Python 物件，混進數值運算裡會出問題。
```

```python
long_zip_codes = requests['Incident Zip'].str.len() > 5
requests['Incident Zip'][long_zip_codes].unique()
```

```python
# 有些是 10038-1234 這種格式，只取前五碼
requests['Incident Zip'] = requests['Incident Zip'].str.slice(0, 5)
requests['Incident Zip'].unique()
```

```python
# 紐約的郵遞區號應該在 10000-11356 之間
zips = requests['Incident Zip'].dropna()
is_valid = zips.astype(float).between(10000, 11356)
zips[~is_valid].unique()
```
